```python
import cmaboss, pandas, maboss, json, collections
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### MaBoSS

```python
sim = maboss.load("p53/p53_toy.bnd", "p53/p53_toy.cfg")
```

```python
res = sim.run()
```

```python
fig, ax = plt.subplots(1,1,figsize=(4,3),dpi=500)
res.get_states_probtraj().plot.area(ax=ax)
plt.legend(bbox_to_anchor=(1.1, 0.3), loc='upper left')
plt.xlim(0,5)
plt.ylim(0,1)
plt.xlabel("time")
plt.ylabel("probabilities")
```

### PopMaBoSS

```python
sim = maboss.PopSimulation("p53/p53_toy_pmb.pbnd", "p53/p53_toy_pmb.cfg")
```

```python
sim.param["sample_count"] = 1000
```

```python
res = sim.run()
```

```python
fig, ax = plt.subplots(1,1,figsize=(4,3),dpi=500)
res.get_simple_states_probtraj().plot.area(ax=ax)
plt.xlim(0,5)
plt.ylim(0,1)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel("time")
plt.ylabel("probabilities")
```

```python
fig, ax = plt.subplots(1,1,figsize=(4,3),dpi=500)
res.get_simple_popsize().plot(ax=ax)
# plt.xlim(0,5)
# plt.ylim(0,100)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel("time")
plt.ylabel("population size")

pd.DataFrame(
    [
        res.get_simple_popsize()-(res.get_simple_popsize_errors()**0.5), res.get_simple_popsize()+(res.get_simple_popsize_errors()**0.5), 
    ], 
).transpose().plot(
    color=['C0', 'C0'],
    logy=True,
    legend=False,
    ax=ax, style='-.'
)
```

```python
fig, ax = plt.subplots(1,1,figsize=(4,3),dpi=500)
res.get_simple_states_popsize().plot.area(ax=ax)
# plt.xlim(0,5)
# plt.ylim(0,60)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel("time")
plt.ylabel("population sizes")
```
