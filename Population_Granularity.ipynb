```python
import numpy as np
import math
```

```python
x = range(1, 100)
bases = [2]#1.1, 1.3, 1.5, 2]
xs = {}

for base in bases:

    levels = {}
    for t_x in x:
        level = math.pow(base, round(math.log(t_x)/math.log(base)))
        if level not in levels:
            levels[level] = [t_x]
        else:
            levels[level].append(t_x)

    means = {}
    replace = {}
    print(levels)
    for level, values in levels.items():
        
        means[level] = np.mean(values)
        # print(means)
        for value in values:
            replace[value] = means[level]
        
    xs[base] = [replace[t_x] for t_x in x]

fig, ax = plt.subplots(1,3, figsize=(20,6))

ax[0].plot(x[0:10],x[0:10])
for base in bases:
    ax[0].plot(x[0:10], xs[base][0:10], label=base)
ax[0].legend()

ax[1].plot(x[0:100],x[0:100])
for base in bases:
    ax[1].plot(x[0:100], xs[base][0:100], label=base)
ax[1].legend()

ax[2].plot(x,x)
for base in bases:
    ax[2].plot(x, xs[base], label=base)
ax[2].legend()
```
